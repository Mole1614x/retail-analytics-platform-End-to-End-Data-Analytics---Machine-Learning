```python

P_A = 0.7      
P_B = 0.3      

P_D_given_A = 0.04   
P_D_given_B = 0.05   


P_D = (P_A * P_D_given_A) + (P_B * P_D_given_B)


P_A_given_D = (P_D_given_A * P_A) / P_D


print("Probability that defective item was produced by Machine A:")
print(round(P_A_given_D, 4))
```

```python
p_d = 0.02                 
p_not_d = 0.98             

p_t_given_d = 0.95            
p_t_given_not_d = 0.10     


p_t = (p_d * p_t_given_d) + (p_not_d * p_t_given_not_d)


p_d_given_t = (p_t_given_d * p_d) / p_t


print(f"Probability that person is actually diseased: {p_d_given_t:.4f}")
```

```python
p_a = 0.5
p_b = 0.3
p_c = 0.2
p_d_given_a = 0.02
p_d_given_b = 0.03
p_d_given_c = 0.05
p_d = (p_a*p_d_given_a)+(p_b*p_d_given_b)+(p_c*p_d_given_c)
p_c_given_d =(p_c*p_d_given_c)/p_d
p_c_given_d
print("p_c_given_d =", round(p_c_given_d, 3))
```
