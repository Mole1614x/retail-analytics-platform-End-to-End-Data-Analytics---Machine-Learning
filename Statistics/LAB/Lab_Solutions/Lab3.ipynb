```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
x = df['sepal_length']
y = df['petal_length']

correlation = x.corr(y)
print("correlation:", correlation)
```

```python
x = df['sepal_length']
y = df['petal_length']

covariance = x.cov(y)

correlation = x.corr(y)

print("Covariance:", covariance)
print("Correlation:", correlation)
```

```python
x = df['sepal_length']
y = df['petal_length']
covariance = np.cov(x,y)
print("covariance Matrix:", covariance)
```

```python
sns.scatterplot(x=x,y=y)
plt.title("Relationship between sepal Length and Petal Length")
plt.show()
```

```python
sns.histplot(data=df,x ='sepal_length', kde=True)
plt.title("Histogram with KDE - Customer Species")
plt.xlabel("sepal_length")
plt.ylabel("Frequency")
plt.show()
```

```python
df1 = pd.DataFrame({
    'x' : [10,20,30,40,1000],
    'y' : [15,25,35,45,1100]
})
covariance = x.cov(y)
print(f"covariance:", covariance)

correlation = x.corr(y)

print("correlation:", correlation)
```

```python
q1 = df1['x'].quantile(0.25)
q3 = df1['y'].quantile(0.75)
iqr = q3-q1
lower = q1-1.5*iqr
upper = q3+1.5*iqr
print(lower,upper)
```

```python
clean_df = df[(df['sepal_length'] >= lower) &
                    (df['sepal_length'] <= upper)]
print("Original Shape:", df.shape)
print("Äfter removing outliers:", clean_df.shape)
```

```python
new_corr = clean_df['sepal_length'].corr(clean_df['petal_length'])
new_cov = clean_df['sepal_length'].cov(clean_df['petal_length'])
print("Correlation after removing outliers:", new_corr)
print("Covariance after removing outliers:", new_cov)
```
