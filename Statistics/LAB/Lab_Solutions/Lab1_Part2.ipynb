```python
df_males = df[df['Gender'] == 'Male'][['Gender', 'Monthly_Salary', 'Age']] \
    .sort_values(
        by=['Monthly_Salary', 'Age'],
        ascending=[True, False]
    )
df_males.head()
```

```python
df_female = df[df['Gender'] == 'Female'][['Gender', 'Monthly_Salary', 'Age']] \
    .sort_values(
        by=['Monthly_Salary', 'Age'],
        ascending=[True, False]
    )

df_female.head()
```

```python
df_salary = pd.DataFrame({
    'EmployeeID': df['EmployeeID'],
    'Gross_Salary': df['Monthly_Salary'] * 12,
    'Tax_Paid': (df['Monthly_Salary'] * 12) * 0.20,
    'Net_Salary': (df['Monthly_Salary'] * 12) - ((df['Monthly_Salary'] * 12) * 0.20),
    'Next_Year_Salary': ((df['Monthly_Salary'] * 12) - ((df['Monthly_Salary'] * 12) * 0.20)) * 0.30
})

df_salary.head()
```

```python
df_dept = df.groupby(['Department', 'Gender'])['Monthly_Salary'] \
            .sum() \
            .reset_index() \
            .sort_values(by=['Monthly_Salary'], ascending=True)

df_dept.head()
```

```python
df_edu = df.groupby('Education_Level')[['Age', 'Monthly_Salary']] \
            .sum() \
            .reset_index() \
            .sort_values(by=['Age', 'Monthly_Salary'], ascending=[True, False])

df_edu
```
