```python
df1 = sns.load_dataset('iris')
```

```python
p_setosa = len(df1[df1['species'] == 'setosa'])/len(df1)
p_petal_given_setosa = len(
    df1[(df1['species'] == 'setosa') & (df1['petal_length'] > 1.4)]
) / len(
    df1[df1['species'] == 'setosa']
)
p_petal = len(df1[df1['petal_length'] > 1.4])/len(df1)

prob_a = (p_petal_given_setosa*p_setosa)/p_petal
print(f"Setosa | Petal Length > 1.4: {round(prob_a, 2)}")
                                                        
```

```python
p_setosa = len(df1[df1['species'] == 'setosa']) / len(df1)
p_petal_given_setosa = len(
    df1[
        (df1['species'] == 'setosa') &
        (df1['petal_length'] > 1.4)
    ]
) / len(df1[df1['species'] == 'setosa'])
p_petal = len(
    df1[df1['petal_length'] > 1.4]
) / len(df1)
prob_a = (p_petal_given_setosa * p_setosa) / p_petal

print(f"P(Setosa | Petal Length > 1.4) = {round(prob_a, 2)}")
```
